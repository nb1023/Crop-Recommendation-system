```python
df_combined['EC_correct'] = df_combined['EC'] * 100

df_combined['EC'] =df_combined['EC_correct']

df_combined.drop(columns='EC_correct', inplace=True)
df_combined.head(10)
```

```python
def classify_n(n):
    if n < 240: return 1
    elif n <= 500: return 2
    else: return 3

def classify_p(p):
    if p < 20: return 1
    elif p <= 50: return 2
    else: return 3

def classify_k(k):
    if k < 125: return 1
    elif k <= 300: return 2
    else: return 3

def classify_oc(oc):
    if oc < 0.5: return 1
    elif oc <= 0.75: return 2
    else: return 3

def classify_ph(ph):
    if ph < 6.5: return 1
    elif ph <= 7.5: return 2
    else: return 3

def classify_ec(ec):
    if ec < 8: return 1
    elif ec <= 16: return 2
    else: return 3
```

```python
df_combined['N_rating'] = df_combined['Nitrogen'].apply(classify_n)
df_combined['P_rating'] = df_combined['Phosphorus'].apply(classify_p)
df_combined['K_rating'] = df_combined['Potassium'].apply(classify_k)
df_combined['OC_rating'] = df_combined['Carbon'].apply(classify_oc)
df_combined['pH_rating'] = df_combined['pH'].apply(classify_ph)
df_combined['EC_rating'] = df_combined['EC'].apply(classify_ec)

# Calculate Soil Fertility Index (SFI)
df_combined['SFI'] = df_combined[['N_rating','P_rating','K_rating','OC_rating','pH_rating','EC_rating']].mean(axis=1)

# Classify Soil Fertility Level
def classify_sfi(sfi):
    if sfi < 1.67: return 'Low'
    elif sfi <= 2.33: return 'Medium'
    else: return 'High'
```

```python
df_combined['Fertility_Level'] = df_combined['SFI'].apply(classify_sfi)
```

```python
columns_to_drop = ['N_rating', 'P_rating', 'K_rating', 'OC_rating', 
                   'pH_rating', 'EC_rating']
df_combined_cleaned = df_combined.drop(columns=columns_to_drop)
```

```python
df_combined_cleaned.head()
```

```python
reference_crops = pd.DataFrame({
    'Crop': ['Rice', 'Wheat', 'Sorghum', 'Maize', 'Sugarcane', 'Cotton', 'Soyabean', 'Tomato', 'Potato', 'Onion', 'Bean'],
    'Available_Nitrogen': [274.4, 338.69, 301.06, 301.06, 225.79, 150.53, 150.53, 163.07, 163.07, 163.07, 225.97],
    'Available_Phosphorus': [15.5, 10.32, 19.78, 19.78, 24.08, 11.47, 11.47, 18.64, 18.64, 18.64, 22.8],
    'pH': [6.2, 6.61, 7.37, 7.37, 8.06, 7.35, 7.35, 8.06, 8.06, 8.06, 7.54],
    'Fertility_Level': [2, 2, 1, 1, 2, 1, 2, 0, 2, 1, 2]
})

reference_crops.head(11)
```

```python
from math import sqrt

def get_ideal_crop(n, p, ph, fertility_num):
    best_crop = None
    min_dist = float('inf')
    
    for _, ref in reference_crops.iterrows():
        if ref["Fertility_Level"] != fertility_num:
            continue
        dist = sqrt((n - ref["Available_Nitrogen"])**2 + (p - ref["Available_Phosphorus"])**2 + (ph - ref["pH"])**2)
        if dist < min_dist:
            min_dist = dist
            best_crop = ref["Crop"]
    
    return best_crop
```

```python
fertility_map = {'Low': 0, 'Medium': 1, 'High': 2}
df_combined_cleaned['Fertility_Level_Num'] = df_combined_cleaned['Fertility_Level'].map(fertility_map)
```

```python
df_combined_cleaned['Ideal_Crop'] = df_combined_cleaned.apply(
    lambda row: get_ideal_crop(row['Nitrogen'], row['Phosphorus'], row['pH'], row['Fertility_Level_Num']),
    axis=1
)
```
